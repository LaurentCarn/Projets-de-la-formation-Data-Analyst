### Données Bilan Alimentaire

Ces données concernent tous types de produits alimentaires. Ne seront conservées que les données relatives aux viandes de volaille pour le cadre de l'analyse.

```python
food_balances.head(100)
```

```python
food_col = food_balances.columns.tolist()
print(food_col)
```

```python
# suppression des colonnes non pertinentes
df_food_balances = food_balances.drop(['Code Domaine','Domaine','Année','Code zone','Code Élément','Code Produit','Code année',
                                            'Symbole','Description du Symbole'], axis=1)
```

```python
df_food_balances.head()
```

```python
prod_list = df_food_balances.Produit.unique().tolist()
print(prod_list)
```

```python
# constitution d'un dataframe avec uniquement les viandes de volailles
df_volaille = df_food_balances[df_food_balances['Produit'] == 'Viande de Volailles']
df_volaille.head()
```

```python
# création d'un attribut "élément + unité" pour faciliter le traitement du pivot
df_volaille = df_volaille.copy()
df_volaille['item_unite'] = df_volaille['Élément'].map(str) + '-' + df_volaille['Unité'].map(str)
```

```python
df_volaille.head(50)
```

```python
# pivot pour avoir un dataframe des éléments de bilan alimentaire par pays
df_volaille = df_volaille.pivot(index='Zone', columns = 'item_unite', values = 'Valeur')
df_volaille.fillna(value = 0, axis =1, inplace = True)
```

```python
df_volaille.head()
```

```python
df_volaille.columns.values
```

```python
len(df_volaille)
```

```python
df_volaille.describe()
```

```python
# suppression des données non pertinentes dans le cadre d'un project d'export de viande de volaille destinée à la
# nourriture humaine ou redondantes avec d'autres données.
df_volaille = df_volaille.drop([('Alimentation pour touristes-Milliers de tonnes'),
                              ('Aliments pour animaux-Milliers de tonnes'),
                              ('Autres utilisations (non alimentaire)-Milliers de tonnes'),
                              ('Semences-Milliers de tonnes'), ('Traitement-Milliers de tonnes'),
                              ('Résidus-Milliers de tonnes'),('Variation de stock-Milliers de tonnes'),
                               'Disponibilité alimentaire (Kcal/personne/jour)-Kcal/personne/jour',
                               'Disponibilité de matière grasse en quantité (g/personne/jour)-g/personne/jour',
                               'Pertes-Milliers de tonnes'], axis = 1)
```

```python
df_volaille.head()
```

```python
# epxression des données en tonnes (pl)
for col in ['Disponibilité intérieure-Milliers de tonnes',
       'Exportations - Quantité-Milliers de tonnes',
       'Importations - Quantité-Milliers de tonnes',
       'Nourriture-Milliers de tonnes',
       'Production-Milliers de tonnes']:
    df_volaille[col] = df_volaille[col] * 1000
    

df_volaille.head()
```

```python
# optimisation des libellés de colonnes
# dictionnaire avec clés = libellés à remplacer et valeurs = nouveaux libellés

dict_libelles_vol = {
                'Disponibilité alimentaire (Kcal/personne/jour)-Kcal/personne/jour':'dispo_alim_kcal_jr_pers',
                'Disponibilité alimentaire en quantité (kg/personne/an)-kg':'dispo_alim_kg_an_pers',
                'Disponibilité de matière grasse en quantité (g/personne/jour)-g/personne/jour':'dispo_mat_grasse_g_jr_pers',
                'Disponibilité de protéines en quantité (g/personne/jour)-g/personne/jour':'dispo_prot_g_jr_pers',
                'Disponibilité intérieure-Milliers de tonnes':'dispo_interieure',
                'Exportations - Quantité-Milliers de tonnes':'exportations',
                'Importations - Quantité-Milliers de tonnes':'importations',
                'Nourriture-Milliers de tonnes':'nourriture',
                'Pertes-Milliers de tonnes':'pertes',
                'Production-Milliers de tonnes':'production'}

# utilisation liste en compréhension pour recherche et remplacement

df_volaille.columns = [dict_libelles_vol.get(x, x) for x in df_volaille.columns]
df_volaille.head()
```

### ACP

```python
# tranformation des données en numpy array

X = df_pca.values
X.shape
```

```python
# réduction et centrage des données :

from sklearn import preprocessing

std_scale = preprocessing.StandardScaler().fit(X)
X_scaled = std_scale.transform(X)
```

```python
# Calcul des composantes principales

from sklearn import decomposition

pca = decomposition.PCA()
pca.fit(X_scaled)
```

### Eboulis des valeurs propres

```python
# Eboulis des valeurs propres + courbe de l'inertie cumulée

plt.rcParams["figure.figsize"] = [12,7]
display_scree_plot(pca)
```

```python
# calcul du % de variance expliquée par chacune des composantes
# pca.explained_variance_ratio_

print('% de variance expliquée pour chaque composante : \n',pca.explained_variance_ratio_)
print('la somme des % de variance expliquée vaut : \n',pca.explained_variance_ratio_.sum())
print('Sommes cumulées des % de variance expliquée : \n',pca.explained_variance_ratio_.cumsum())
```

**Constat**:<br>

46 % de la variance est expliquée par les 2 premières composantes, soit 20% des variables synthétiques.

### Cercle des corrélations sur le premier plan factoriel

```python
# pour mieux comprendre ce que capturent les deux composantes principales, on utilise pca.components_
# qui nous donne les coordonnées des composantes principales dans l'espace intial(dimension = nb de variables)

## on affiche pour chacune des variables, un point dont l'abscisse sera sa contribution à la première PC
## et l'ordonnée sa contribution à la deuxième

pcs = pca.components_

fig = plt.figure(figsize = [10,10])

for i, (x, y) in enumerate(zip(pcs[0,:], pcs[1,:])):
    # afficher un segment de l'origine au point (x,y)
    plt.plot([0,x], [0,y], color = 'k')
    # afficher le nom (data.columns[i]) de la performance
    plt.text(x, y, df_pca.columns[i], fontsize = '14')
    
# afficher la ligne horizontale y=0

plt.plot([-0.7, 0.7], [0,0], color = 'grey', ls='--')

# afficher la ligne verticale x=0

plt.plot([0,0], [-0.7, 0.7], color = 'grey', ls='--')
plt.xlim([-0.3, 0.7])
plt.ylim([-0.7, 0.5])
plt.title('Projection des variables sur le 1er plan factoriel')
plt.xlabel('F{} ({}%)'.format(0+1, round(100*pca.explained_variance_ratio_[0],1)))
plt.ylabel('F{} ({}%)'.format(1+1, round(100*pca.explained_variance_ratio_[1],1)))

plt.show()
```

### Projection des pays sur le premier plan factoriel

```python
# on peut représenter chaque pays selon ces deux dimensions uniquement

## projection de X sur les composantes principales

X_projected = pca.transform(X_scaled)

## afficher chaque observation

fig = plt.figure(figsize = [15,10])
sns.scatterplot (x=X_projected[:,0], y=X_projected[:,1],
            ## coloration en fct de la proportion de protéines de volailles dans le total des protéines animales dispo
            hue=df_pca.get('production'), palette = 'crest',
            ## taille en fonction des volumes d'importationde viandes de volailles
            size=df_pca.get('importations'), sizes=(80, 200))

for i in range(df_pca.shape[0]):
    plt.text(X_projected[i,0], X_projected[i,1],df_pca.index[i])

plt.plot([-4, 5], [0, 0], color='grey', ls='--')
plt.plot([0, 0], [-3, 6], color='grey', ls='--')
plt.title('Projection des pays sur le 1er plan factoriel')
plt.xlabel('F{} ({}%)'.format(0+1, round(100*pca.explained_variance_ratio_[0],1)))
plt.ylabel('F{} ({}%)'.format(1+1, round(100*pca.explained_variance_ratio_[1],1)))

plt.show()
```

## Classification Ascendante Hierarchique

### Dendrogramme complet

```python
# Clustering hiérarchique
from scipy.cluster.hierarchy import linkage, fcluster

# sur X_projected, résultat de l'ACP
Z = linkage(X_projected, 'ward', metric = 'euclidean', optimal_ordering = True)

# Affichage du dendrogramme

plt.figure(figsize=(15,25))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('distance')
dendrogram(
    Z,
    labels = df_pca.index,
    orientation = "left",
    color_threshold = 15
)
plt.show()
```

## Kmeans

### Métrique silouhette pour indication sur le nombre de clusters optimal

```python
#utilisation de la métrique "silhouette"
#faire varier le nombre de clusters de 2 à 10
from sklearn.cluster import KMeans
from sklearn import decomposition
from sklearn import metrics
res = np.arange(9,dtype="double")
for k in np.arange(9):
    km_sil = KMeans(n_clusters=k+2)
    km_sil.fit(X)
    res[k] = metrics.silhouette_score(X,km_sil.labels_)
print(res)
#graphique
import matplotlib.pyplot as plt
fig = plt.figure(figsize = [10,6])
plt.title("Silhouette")
plt.xlabel("# of clusters")
plt.plot(np.arange(2,11,1),res)
plt.show()
```

Le coefficient de silhouette est optimal avec 2 clusters.<br>

Toutefois, ce coefficient reste très satisfaisant sur la plage entre **3 et 5 clusters**.<br>

Ce qui permet de choisir un clustering en **4 groupes**, et ainsi de comparer facilement les résultats du K-means avec ceux de la CAH.

### Clustering

```python
from sklearn.cluster import KMeans
from sklearn import decomposition

# Nombre de clusters souhaités
n_clust = 4


# X_scaled = preprocessing.StandardScaler().fit_transform(X)

# Clustering par K-means
km = KMeans(n_clusters=n_clust, random_state = 26, n_init = 80)
km.fit(X_projected)

# Récupération des clusters attribués à chaque individu
km_clusters = km.labels_
```
